## Question 1: What is Simple Linear Regression (SLR)? Explain its purpose.

Simple Linear Regression (SLR) is a supervised learning technique that models the linear relationship between one independent variable and one dependent variable.  
It is used for prediction and understanding variable relationships.

---

## Question 2: Key assumptions of Simple Linear Regression

1. Linear relationship  
2. Independent observations  
3. Homoscedasticity  
4. Normality of residuals  
5. No multicollinearity  

---

## Question 3: Mathematical equation

y = β₀ + β₁x + ε  

- β₀: Intercept  
- β₁: Slope  
- ε: Error term

## Question 5: Method of least squares

## Question 10: Interpretation of coefficients

## Question 9: Python Code — Simple Linear Regression using scikit-learn

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 4, 6, 8, 10])

# Model
model = LinearRegression()
model.fit(X, y)

print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)
```
